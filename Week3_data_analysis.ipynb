```python
import pandas as pd
import matplotlib.pyplot as plt
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

import import_ipynb
from Week1_data_analysis_season import summer_events, winter_events
```

### Beyond Descriptive  Statistical

```python
#Summer Olympics:
summer_medal_count = pysqldf('''
                         SELECT
                             Year,
                             COUNT(*) AS total_count,
                             SUM(CASE
                                   WHEN Medal IS NOT NULL THEN 1 ELSE 0
                                 END) AS medal_count,
                             SUM(CASE
                                   WHEN Medal = "Gold" THEN 1 ELSE 0
                                 END) AS gold_count,
                             SUM(CASE
                                   WHEN Medal = "Silver" THEN 1 ELSE 0
                                 END) AS silver_count,
                             SUM(CASE
                                   WHEN Medal = "Bronze" THEN 1 ELSE 0
                                 END) AS bronze_count
                             FROM
                               summer_events
                             GROUP BY
                               Year 
                                   ''')
```

```python
#Winter Olympics:
winter_medal_count = pysqldf('''
                         SELECT
                             Year,
                             COUNT(*) AS total_count,
                             SUM(CASE
                                   WHEN Medal IS NOT NULL THEN 1 ELSE 0
                                 END) AS medal_count,
                             SUM(CASE
                                   WHEN Medal = "Gold" THEN 1 ELSE 0
                                 END) AS gold_count,
                             SUM(CASE
                                   WHEN Medal = "Silver" THEN 1 ELSE 0
                                 END) AS silver_count,
                             SUM(CASE
                                   WHEN Medal = "Bronze" THEN 1 ELSE 0
                                 END) AS bronze_count
                             FROM
                               winter_events
                             GROUP BY
                               Year 
                                   ''')
```

**I have generated two tables that tally the overall count of medals awarded in the winter and summer Olympics, respectively. My next step is to compute the Pearson correlation coefficient between these two sets of data.**

The Summer Olympics began in 1896 and the Winter Olympics started in 1924, the number of medal counts in their respective arrays are different. As a result, I need to generate a truncated table for the Summer Olympics that only includes data from 1924 onwards to match the length of the Winter Olympics array.

```python
# I exclude the first 7 rows (which correspond to the years before 1924)
summer_medal_count_new = summer_medal_count[7:]
```

```python
summer_medal_count_new
```

Next, I compute the Pearson correlation coefficient for the overall medal counts of the Winter and Summer Olympics between the years 1924 and 2016.

```python
import numpy as np
```

```python
x_simple = winter_medal_count.medal_count
y_simple = summer_medal_count_new.medal_count
coffi = np.corrcoef(x_simple, y_simple)
```

```python
coffi
```

The Pearson correlation coefficient between the overall medal counts of the Winter and Summer Olympics from 1924 to 2016 is 0.94, indicating a strong positive correlation. This suggests that a country's performance in the Winter Olympics is highly linked to its performance in the Summer Olympics.

**Next, I will compute the standard deviation in country performance over time. To assess the difference, I will compare the mean standard deviation of the Winter Olympics to that of the Summer Olympics.**

```python
std_medal_count_summer_olympics = np.std(y_simple)
std_medal_count_winter_olympics = np.std(x_simple)
```

```python
print("std_medal_count_summer_olympics =",std_medal_count_summer_olympics)
print("std_medal_count_winter_olympics =",std_medal_count_winter_olympics)
```

Between 1924 and 2016, the standard deviation in medal count for the Summer Olympics was approximately three times greater than that of the Winter Olympics. This indicates that a country's performance varies more from year to year in the Summer Olympics.
